```python
def get_forecast_data(station):
    request_fc.all().df.to_pandas()
    stations = request_fc.filter_by_station_id(
            station_id=[station],
        )

    x = stations.values.all().df.to_pandas()
    print(x.parameter.unique())
    transformed_data = x.pivot_table(index='date', columns='parameter', values='value', aggfunc='first')
    transformed_data.head(24).plot()
    return transformed_data.head(24)

fc_station_list =["10379",
                  "10505",
                  "10438",
                  "10729"]
```
